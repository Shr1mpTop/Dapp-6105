```python
import sqlite3
import datetime
conn = sqlite3.connect('user.db')
conn.execute('CREATE TABLE IF NOT EXISTS user (name text, timestamp timestamp)')
conn.close()
```

# 创建用户表

创建 SQLite 数据库中的用户表，包含 name 和 timestamp 字段。

```python
#check
conn = sqlite3.connect('user.db')
c = conn.cursor()
c.execute('select * from user')
print(c.description)
c.close
conn.close()
```

# 检查用户表结构

查询用户表的列描述信息。

```python
sqlite3.register_adapter(datetime.datetime, lambda dt: dt.isoformat())
sqlite3.register_converter("timestamp", lambda s: datetime.datetime.fromisoformat(s))
```
